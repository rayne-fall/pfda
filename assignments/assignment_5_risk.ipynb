This assignment simulates 1000 battle rounds of the game Risk and plots the outcome.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# dice rolls for attackers
x = np.random.uniform(low=1, high=6, size=2000)

# dice rolls for defenders
y = np.random.uniform(low=1, high=6, size=2000)

attack_wins = x>y
defence_wins = x<=y

a_wins_count = attack_wins.sum()
d_wins_count = defence_wins.sum()

labels = ["Attackers Won", "Defenders Won"]
plt.pie(np.array([a_wins_count, d_wins_count]), labels = labels)
```
