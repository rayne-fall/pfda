This week's assignment is to take the CSV file provided in the PFDA-courseware repository and creat a Jupyter notebook containing a plot of the temperature of the dry bulbs over time.

## Imports

I will be importing three modules for this task: PANDAS, NumPy and matplotlib.

```python
# import necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the data

```python
# read in data
df = pd.read_csv("https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/refs/heads/main/assignment/weatherreadings1.csv")
```

# Plotting temperature versus time

```python
# create plot
fig, ax = plt.subplots(figsize = (275, 100))

# get variables
temp = df["dryBulbTemperature_Celsius"]
time = df["reportEndDateTime"]
# convert data frames to arrays
temp = temp.to_numpy()
time = time.to_numpy()

# plot variables
ax.plot(time, temp, "*-", linewidth=2)
# label axes
ax.set_xlabel("Time")
ax.set_ylabel("Temperature (C)")
# add gridlines
ax.grid()
# show plot
plt.show
```
