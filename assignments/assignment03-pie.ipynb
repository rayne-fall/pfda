This week's assignment is to create a pie chart of the email domains in the provided csv file.

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Reading in the data frame

```python
df = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
```

### Extracting the domains from the email addresses

```python
df['domain'] = df['Email'].str.split('@').str[1]
```

### Finding out which domains we have and how many of each there are

```python
df['domain'].value_counts()
```

### Creating the pie chart

```python
fig, ax = plt.subplots()

# give the domains as an array
domains = np.array(["example.org", "example.com", "example.net"])

# get the number of each domain
count = df["domain"].value_counts()

# convert the data frame to an array
count = count.to_numpy() 

# create labels for each section
labels = "example.org", "example.com", "example.net"

# generate pie chart with percentages labelled
ax.pie(count, autopct='%1.1f%%')

# add label and title
ax.legend(labels, bbox_to_anchor=(0.75, 0, 0.5, 1))
ax.set_title('Email Domains')
```
